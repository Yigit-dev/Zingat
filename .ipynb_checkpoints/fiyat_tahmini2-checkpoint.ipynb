```python
def tarih_fonksiyonu(data):
   
    data = data.copy()
    data['yıl'], data['ay'], data['gun'] = data["tarih"].str.split("-",2).str
    data["yıl"] = data["yıl"].astype(str).astype(int)
    data["ay"] = data["ay"].astype(str).astype(int)
    data["gun"] = data["gun"].astype(str).astype(int)
    
    return data
```

```python
trainset = tarih_fonksiyonu(trainset)
```

```python
def belirtilmemis_fonksiyonu(data):
    
    data = data.copy()
    data = data.replace("-","belirtilmemis")
    
    return data
```

```python
trainset = belirtilmemis_fonksiyonu(trainset)
```

```python
def path_split_fonksiyonu(data):
    
    data = data.copy()
    data["il"], data["ilce"], data["mahalle"] = data["path"].str.split("/",2).str
    data["il"] = data["il"].astype(str).astype(int)
    data["ilce"] = data["ilce"].astype(str).astype(int)
    data["mahalle"] = data["mahalle"].astype(str).astype(int)
    
    return data
```

```python
trainset = path_split_fonksiyonu(trainset)
```

```python
def güncel_fiyat_fonksiyonu(data):
    
    data = data.copy()
    data["güncel_fiyat"] = data["güncel_fiyat"].replace(" TRY","")
    data["güncel_fiyat"] = data["güncel_fiyat"].astype(str).astype(int)
    
    return data
```

```python
trainset = güncel_fiyat_fonksiyonu(trainset)
```

```python
def odasayisi_fonksiyonu(data):
   
    data = data.copy()
    data['odasayisi'], data['salonsayisi'] = data['odasayısı'].str.split("+",1).str
    data["odasayisi"] = data["odasayisi"].astype(str).astype(int)
    data["salonsayisi"] = data["salosayisi"].astype(str).astype(int)
    
    return data
```

```python
trainset = odasayisi_fonksiyonu(trainset)
```

```python
label_encoder_cols = ("il","ilce","mahalle","emlaktipi","binayaşı","bulunduğukat","binadakikatsayısı","isıtmatipi","otopark")
```

```python
from sklearn.preprocessing import LabelEncoder
def label_encoder_fonksiyonu(data,cols):
    
    for col_name in label_encoder_cols:
        labelEncoder = LabelEncoder()
        labelEncoder.fit(data[col_name].values)
        data[col_name] = labelEncoder.transform(data[col_name].values)
        
    return data    
```

```python
trainset = label_encoder_fonksiyonu(trainset,label_encoder_cols)
```
